#### EDA
**Task 5: Data Analysis**

**Univariate Analysis**

```python
# Identifying outliers in price
sns.boxplot(data=data,x='price')
plt.savefig('images/boxplot_price.png')
plt.show()
```

```python
# Price distribution 

sns.histplot(data=data, x= 'price')
plt.savefig('price_hist_full.png')
plt.show()
```

```python
#Dealing with outliers
df = data[data['price']<1500]
sns.boxplot(data=df, x= 'price')
plt.savefig('images/price_hist_filtered.png')
plt.show()
```

```python
plt.figure(figsize=(11,8))
sns.histplot(data=df,x='price',bins=100)
plt.title('Price_distribution')
plt.ylabel("Frequency")
plt.show()
```

```python
sns.histplot(data=df, x = 'availability_365')
plt.title('Availability_365 Distribution')
plt.ylabel('Frequency')
plt.figure(figsize=(6,3))
plt.show()
```

```python
df.dtypes
```

```python
# Creating new column 'Price per bed'
df['price per bed']= df['price']/df['beds']
```

```python
#average price per bed
df.groupby(by='neighbourhood_group')['price per bed'].mean()
```

**Bivariate Analysis**

One variable dependency in another variable

```python
df.columns
```

```python
#price dependency on neighbourhood
sns.barplot(data=df,x='neighbourhood_group', y = 'price', hue = 'room_type')
plt.savefig('images/roomtype_distribution.png')
plt.show()
```

```python
# number of reviews and price relation
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='number_of_reviews', y='price', hue='neighbourhood_group')
plt.savefig('images/locality_review_dependency.png')
plt.title('Locality and Review Dependency')
plt.show()
```

```python
df.dtypes
```

```python
sns.pairplot(data=df, vars=['price','minimum_nights', 'number_of_reviews', 'availability_365'], hue = 'room_type')
plt.savefig('images/pairplot.png')
plt.show()
```

```python
#Geographical Distribution of AirBnB
plt.figure(figsize=(11,9))
sns.scatterplot(data=df, x='longitude', y='latitude', hue= 'room_type')
plt.title('Geographical Distribution of AirBnB')
plt.show()
```

```python
# creating heatmap to get correlation of one numerical variable with others
corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr

sns.heatmap(data=corr, annot=True)
plt.savefig('images/borough_price_heatmap.png')
plt.figure(figsize=(10,7))
plt.show()
```
